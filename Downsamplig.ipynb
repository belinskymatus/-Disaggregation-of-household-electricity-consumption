```python
import pandas as pd
import os

# Cesta k priečinku so súbormi
input_folder = ""
output_folder = ""

# Pre každý súbor od 0 po 6
for i in range(7):
    filename = f"redd_house1_{i}.csv"
    input_path = os.path.join(input_folder, filename)

    # Kontrola, či súbor existuje
    if not os.path.isfile(input_path):
        print(f"Súbor {filename} neexistuje, preskakujem.")
        continue

    # Načítanie dát
    df = pd.read_csv(input_path)

    # Odstránenie nepotrebného stĺpca
    if "Unnamed: 0" in df.columns:
        df.drop(columns=["Unnamed: 0"], inplace=True)

    # Pridanie simulovanej časovej osi
    df["timestamp"] = pd.date_range(start="2025-01-01", periods=len(df), freq="s")

    # Nastavenie timestampu ako index a resampling
    df.set_index("timestamp", inplace=True)
    df_resampled = df.resample("5S").mean().dropna()

    # Obnovenie timestampu ako čitateľného stringu
    df_resampled["timestamp"] = df_resampled.index.strftime("%Y-%m-%d %H:%M:%S")

    # Reset indexu
    df_resampled.reset_index(drop=True, inplace=True)

    # Uloženie
    output_filename = f"standardized_house1_{i}.csv"
    output_path = os.path.join(output_folder, output_filename)
    df_resampled.to_csv(output_path, index=False)
    print(f"Spracovaný súbor: {output_filename}")
```
